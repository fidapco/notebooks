## DataCommons Tables

- datacommons_states contains the aggregated data w.r.t states of U.S
- datacommons_counties contains the aggregated data w.r.t counties of U.S
- datacommons_cities contains the aggregated data w.r.t cities of U.S
- datacommons_zipcodes contains the aggregrated w.r.tabout geooverlaps of U.S

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from fidap import fidap_client
import config
```

```python
fidap = fidap_client(api_key=config.FIDAP_API_KEY)
```

# datacommons_states

##### Employment Rate in State 

```python
emp_rate = fidap.sql('select name, place, (employed_persons / population) * 100 as employment_rate from datacommons.states order by employment_rate desc')
pb=emp_rate.plot.bar(x='name', y='employment_rate', ylabel='Employment %', xlabel='States', rot=90, figsize=(22,5))
```

##### GDP per State

```python
male_per=fidap.sql('select name, economic_GDP_value from datacommons.states order by economic_GDP_value;')
ax = male_per.plot(kind='line', y="economic_GDP_value", x="name", figsize=(22,5))
```

##### Asian population per state

```python
fidap.sql('select name, (asian_persons / population)*100 as asian_population_pct, population from datacommons.states order by asian_persons desc limit 1;')
```

#### Asian Population w.r.t population

```python
fidap.sql('select name, (asian_persons / population)*100 as asian_population_pct, asian_persons, population from datacommons.states order by asian_population_pct desc')
```

##### Employment Rate

```python
fidap.sql('select name, (employed_persons / population) * 100 as employment_pct, population from datacommons.states order by employment_pct desc;')
```

##### covid_19 positive test by states %
