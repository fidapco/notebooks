```python
from fidap import fidap_client
import config
```

```python
fidap = fidap_client(api_key=config.FIDAP_API_KEY)
```

```python
fidap.sql("""select table_name, row_count, column_count from meta_tables""")
```

```python
fidap.sql("""select * from meta_fields where "table"='insider_trades' """).head()
```

```python
fidap.sql("""select tickers.ticker, tickers.name,  tickers.industry, tickers.rev_yoy, tickers.pc_1y
from tickers where tickers.sector = 'Technology' and tickers.marketcap > 10e9 and tickers.rev_yoy < 0 
order by pc_1y desc limit 5""")
```
